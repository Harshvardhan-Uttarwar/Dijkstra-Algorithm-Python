```python
def dijkstra(G, start):
    """
    Implementation of dijkstra using adjacency matrix.
    This returns an array containing the length of the shortest path from the start node to each other node.
    It is only guaranteed to return correct results if there are no negative edges in the graph. Positive cycles are fine.
    This has a runtime of O(|V|^2) (|V| = number of Nodes), for a faster implementation see @see ../fast/Dijkstra.java (using adjacency lists)

    :param graph: an adjacency-matrix-representation of the graph where (x,y) is the weight of the edge or 0 if there is no edge.
    :param start: the node to start from.
    :return: an array containing the shortest distances from the given start node to each other node
    """

    print(G)
    # This contains the distances from the start node to all other nodes
    distances = [G.iloc[start,_] for _ in range(len(G))]
    print(distances)

    # This contains whether a node was already visited
    visited = [False for _ in range(len(G))]


    # The distance from the start node to itself is 0
    distances[start] = 0
    path = {v:[]for v in G}
    path[start] = [start]
    count = 0

    for i in G.index:
      for j in G.index:
        min_value = min(G[i][j],G[j][i])
        G[i][j] = min_value
        G[j][i] = min_value

    
    # While there are nodes left to visit...
    while True:
        # ... find the node with the currently shortest distance from the start node...
        shortest_distance = 999999999
        shortest_index = -1
 
        for i in range(len(G)):
            # ... by going through all nodes that haven't been visited yet
            
            if distances[i] < shortest_distance and not visited[i]:
                shortest_distance = distances[i]
                shortest_index = i
                #count = count +1
                print("Visiting node " + str(shortest_index) + " with current shortest distance " + str(shortest_distance))

        if shortest_index == -1:
            # There was no node not yet visited --> We are done
            return distances

      
        # ...then, for all neighboring nodes that haven't been visited yet....
        x = list(range(len(G)))
        x.pop(shortest_index)
        print("x - " , x)
        print("Shortest Index - " , shortest_index)
        for i in x:
            # ...if path over this edge is shorter ...Save this path as new shortest path.
            if distances[i] > distances[shortest_index] + G[i][shortest_index]:
              distances[i] = distances[shortest_index] + G[shortest_index][i]
              print("Updating distance of node " + str(i) + " to " + str(distances[i]))
              path[i] = path[shortest_index] + [i]
              
              
        
        # Lastly, we are finished with this node.
        
        visited[shortest_index] = True
        print("Visited nodes: " + str(visited))
        print("Currently lowest distances: " + str(distances))
      
```

```python
# path from u (start) to v(target) with k edges
def shortestPath(graph, u, v, k):
    Inf = 99999999
    
    for i in graph.index:
      for j in graph.index:
        min_value = min(graph[i][j],graph[j][i])
        graph[i][j] = min_value
        graph[j][i] = min_value

# Base cases
    print("start node - ",u ,"final node" ,v ,"initial steps", k)
    if k == 0 and u == v:
        return 0
    if k == 1 and graph[u][v] != Inf:
        return graph[u][v]
    if k <= 0:
        return Inf
 
# Initializing
    res = Inf
 
# Go to all adjacents of u and recur
    for i in range(len(graph)):
        if graph[u][i] != Inf and u != i and v != i:
            rec_res = shortestPath(graph, i, v, k - 1)
            if rec_res != Inf:
                res = min(res, graph[u][i] + rec_res)
                print("least distance -",res)
    return res
```
